```python
import numpy as np
import pandas as pd
import xarray as xr
import ipywidgets as widgets
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import warnings
warnings.filterwarnings('ignore')
# 读取2010年的海温数据（有海温和海温距平两个属性）
dset = xr.open_mfdataset("ersst.v5.2010*.nc").squeeze()
dset
```

```python
date_list = pd.date_range(start='2010-01',end='2010-12',freq='MS')
dset['time'] = date_list
dset
```

```python
dict_variables = {}
for varname, variable in dset.items():
    if len(dset[varname].attrs) > 0:
        dict_variables[dset[varname].attrs['long_name']] = varname
print(dict_variables)
print(list(dict_variables.keys()))
```

```python
# widget for variables
list_variables = list(dict_variables.keys())
select_variable = widgets.Select(
    options=list_variables,
    value=list_variables[0],
    rows=5,
    description='Variable:',
    disabled=False
)

# Widget for date
select_date = widgets.Dropdown(
    options=np.array(dset.time),
    rows=5,
    description='Date:',
    disabled=False
)

# Widget for colorbar
select_colorbar = widgets.Dropdown(
    options=['SST', 'SSTA'],
    rows=2,
    description='Colorbar:',
    disabled=False
)

# generate plot


def generate_plot(date, variable, colorbar):
    d = {'SST': 'jet', 'SSTA': 'bwr'}
    fig = plt.figure(figsize=[18, 8])
    ax = fig.add_subplot(111, projection=ccrs.PlateCarree(central_longitude=0))
    dset[dict_variables[variable]].sel(time=date).plot(cmap=d[colorbar])
    ax.coastlines(color='k', linewidth=1)
    gl = ax.gridlines(draw_labels=True)
    gl.top_labels = False
    gl.right_labels = False
    plt.tick_params(labelsize=25)


interact_plot = widgets.interact(
    generate_plot,
    date=select_date,
    variable=select_variable,
    colorbar=select_colorbar)
```
